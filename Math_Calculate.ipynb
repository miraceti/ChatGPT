```python
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans
    
        
```

```python
string_frac("3")
```

```python
string_frac("2/5")
```

```python
def one_step_mult():
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print(a, "x= ", b)
    ans_in = (input("x = "))
    answer = b/a
    
    if string_frac(ans_in)==answer:
        print("Corrrect! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

```python
one_step_mult()
```

```python
def one_step_add():
    import random
    a = random.randint(-4,10)
    b = random.randint(2,24)
    print("x + ",a, " = ", b)
    ans = float(input("x = "))
    answer = b-a
    
    if ans==answer:
        print("Corrrect! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

```python
one_step_add()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#dimensions
xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax - xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) #windows size
plt.plot([xmin,xmax],[0,0],'b') #axe x bleu
plt.plot([0,0],[ymin,ymax],'b') #axe y bleu

#paramètres des axes
ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("some graph")
ax.grid(True)  #grille
ax.set_xticks(np.arange(xmin,xmax,2)) #graduation axe x increment +2
ax.set_yticks(np.arange(ymin,ymax,2))

y = 2*x + 1
plt.plot(x,y,label='y=2x+1')
plt.plot([4],[6], 'ro', label='point')
plt.plot(x, 3*x, label='steeper line')
plt.legend()


plt.show()
```
